### How to create and save data in CSV,EXCEL,JSON file

```python
data = {
    "Name" : ["John","Doe","Michel","Jackson","Jack","Mic","Taylor","Johnny"],
    "Age" : [30, 40 , 33 , 44 , 27 , 37 , 32 , 35],
    "Salary" : [50000 , 40000, 30000 , 45000 , 43000,35000 , 30000 , 32000]
}


df = pd.DataFrame(data)

# save in excel file
df.to_excel("output.xlsx" , index=False)

# save in csv file
df.to_csv("output.csv" , index=False)

# save in json file
df.to_json("output.json" , index=False)

```

### Descriptive Statistics

```python
print(df.describe())

# what it gives us
# 1) Count => give the total count of each row
# 2) Mean => total average of the data in each row # formula => (sum of data / length of data)
# 3) STD => mean tell us how much distance from mean value
# 4) Min => minimum value
# 5) Max => maximum value
```

```python
# info method tell us about our data set , how much col ,row , null , dtype
print(df.info())
```

### Selecting column and filter row

```python
# first i check my data shape and cols names
print("Cols in my data \n" ,df.columns , "\n")
print("Shape of my data \n" ,df.shape , "\n")



# selecting col
print("Selected col \n", df[["Age" , "Salary"]] , "\n")


# filtration on row
print(df[df["Age"] > 40] , "\n")
print(df[(df["Age"] > 30) & (df["Salary"] >= 40000)])
```
