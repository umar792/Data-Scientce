```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure = [restaurantId , 2021 , 2023 , 2024 ]

sales_Data = np.array([
    [1, 150000 , 180000 , 220000 , 250000],
    [2, 120000 , 140000 , 160000 , 190000],
    [3, 200000 , 240000 , 260000 , 300000],
    [4, 180000 , 210000 , 240000 , 270000],
    [5, 185000 , 160000 , 205000 , 230000],
])
print("shape \n", sales_Data.shape) # first need to know about the shape of data
print("dimension \n", sales_Data.ndim) # second need to find the dimension
print("size \n" , sales_Data.size) # third need to find the data size
```

```python
# find total sales per year
total_yearly_sales = np.sum(sales_Data[: , 1:], axis=0) # everything:everything , start from 1 index : everything
print("Yearly total data \n" , total_yearly_sales , "\n")
total_Sales = np.sum(total_yearly_sales)
print("total sales \n", total_Sales, "\n")

# minimum sales for each
minimum_Sales = np.min(sales_Data[: , 1:], axis=1)
print("minimum sales \n" , minimum_Sales, "\n")

# maximum sales per year
max_Sales_Year = np.max(sales_Data[: , 1:], axis=0)
print("maximum sales per year \n", max_Sales_Year, "\n")


# average sales per restaurant 
avg_sales = np.mean(sales_Data[: , 1:], axis=1)
print("average sales per restaurant \n", avg_sales, "\n")
```

```python
# graph view
plt.figure(figsize=(12,6))
plt.plot(avg_sales)
plt.title("Average sales according to yearly")
plt.xlabel("Yearly")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
