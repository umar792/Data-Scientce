#  Calculate the total world populations 

```python
import numpy as np
import csv
```

```python
data_set = []
with open("./simple_world_population.csv", "r", encoding="utf-8") as file :
    read_file = csv.reader(file)
    for row in read_file:
        data_set.append(row)

remove_header_data_set = data_set[1:]
```

```python
# create np array of remove_header_data_set

data_np = np.array(remove_header_data_set)

# first we will find the shape of the data
print("shape \n", data_np.shape, "\n")

# now find the dimension
print("dimension \n" , data_np.ndim , "\n")

# view the data set how it look like
one_data_set = data_np[0]
print("data set")
print(data_set[0])
print(one_data_set)



```

```python
# find the total population from 2001 to 2025
total_population = np.sum(data_np[:, 2].astype(int), axis=0)
print("total population from 2001 to 2025 \n" , total_population , "\n")

# find total male population
total_male_population = np.sum(data_np[: , 3].astype(int), axis=0)
print("total male population \n" , total_male_population , "\n")

# find total female population
total_female_population = np.sum(data_np[: , 4].astype(int), axis=0)
print("total female population \n" , total_female_population , "\n")

# find total child population
total_child_population = np.sum(data_np[: , 5].astype(int), axis=0)
print("total child population \n" , total_child_population , "\n")

# find each year population
year_array = np.array([[int(row[1]) , int(row[2])] for row in data_np])
unique_year = np.unique(year_array[: , 0])
print(unique_year)


unique_yearly_population = []

for year in unique_year:
    sum = np.sum(year_array[year_array[: , 0] == year][: , 1])
    unique_yearly_population.append(int(sum))
print(unique_yearly_population)


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(unique_year,unique_yearly_population)
plt.title("Total Population Per Year")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
labels = ["Male", "Female", "Chils"]
values = [total_male_population , total_female_population , total_child_population]
colors = ["blue","green","purple"]
plt.bar(labels,values , color=colors)
plt.title("Total Global Population by Gender and Age Group")
plt.ylabel("Population")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
