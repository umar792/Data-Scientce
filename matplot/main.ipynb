# Basics of matplot |

```python
import matplotlib.pyplot as plt
```

```python
months = [1,2,3,4,5,6,7,8,]
sales = [1000, 100, 3000, 500, 1000, 0, 2200,4000]
plt.plot(months, sales , marker='o',linestyle="--",label="Sales",color="blue")
plt.title("Monthly Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid()
```

# Bar chat

```python
products = ['A', 'B', 'C', 'D', 'E']
sales = [1000, 2000, 3000, 4000, 5000]
plt.bar(products , sales, color='orange')
plt.title("Product Sales Data")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.legend()
plt.grid(linestyle=':',linewidth=0.5)
plt.tight_layout()
plt.show()
# The code above creates two plots: a line plot for monthly sales data and a bar plot for product sales data.
```

## Pie chart

```python
regions = ['North', 'South', 'East', 'West']
sales = [1000, 2000, 3000, 4000]
plt.pie(sales, labels=regions, autopct='%1.2f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Sales Distribution by Region")
plt.show()
```

## histogram

```python
scores = [90,91,92,93,94,95,100,50,20,30,40,50,60,70,80,90,10]
plt.hist(scores, bins=3, color='purple', edgecolor='black',)
plt.title("Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

# Scatter"

```python
products = ['A', 'B', 'C', 'D', 'E']
sales = [1000, 2000, 3000, 4500, 500]
plt.scatter(products , sales, color='orange')
plt.title("Product Sales Data")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.legend()
plt.grid(linestyle=':',linewidth=0.5)
plt.tight_layout()
plt.show()
```
